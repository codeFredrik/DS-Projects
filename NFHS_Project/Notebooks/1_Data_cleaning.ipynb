```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

Path = ("C:\\Users\\Fredrik\\vscode_projects\\Projects\\NFHS_Project\\Data\\raw\\india-nfhs4-final.xlsx")

df_raw = pd.read_excel(Path)
print(f"Shape: {df_raw.shape}")
print("\nFirst 10 rows (raw):")
df_raw.head(10)
```

```python
# Row structure
for i in range(6):
    print(f"\n Row {i}:")
    print(df_raw.iloc[i].tolist())
```

```python
Header_row = 2    
Data_row = 3  

# Extracting headers
headers = df_raw.iloc[Header_row].tolist()

# Slicing
df = df_raw.iloc[Data_row:].copy()
df.columns = headers
df = df.reset_index(drop=True)

print(f"Shape after slicing: {df.shape}")
print("\nColumn names:")
for i, col in enumerate(df.columns):
    print(f" [{i}] {col}")
```

```python
import re

def clean_col_name(name):
    """Lowercase, strip, replace spaces/special chars with underscores."""
    if not isinstance(name, str):
        name = str(name)
    name = name.strip().lower()
    name = re.sub(r"[^\w\s]", "", name)       # removes special characters
    name = re.sub(r"\s+", "_", name)          # spaces to underscores
    name = re.sub(r"_+", "_", name)           # collapse multiple underscores
    return name.strip("_")

df.columns = [clean_col_name(c) for c in df.columns]

# Checking for duplicate column names
dupes = df.columns[df.columns.duplicated()].tolist()
if dupes:
    print(f" Duplicate columns found: {len(dupes)}")
    # Deduplicating by appending _1, _2..
    def _dedupes(columns):
        counts = {}
        new_columns = []
        for c in columns:
            if c in counts:
                counts[c] += 1
                new_columns.append(f"{c}_{counts[c]}")
            else:
                counts[c] = 0
                new_columns.append(c)
        return new_columns
    df.columns = _dedupes(df.columns)
else:
    print(" No duplicate column names.")

print(f"\nCleaned column names ({len(df.columns)} total):")
print(df.columns.tolist())
```

```python
Suppressed_val = ["*", "**", "***", "na", "n/a", "N/A", "NA", "", "-", "...", " ", "", "ns", "NS"]

# Replace across entire dataframe
df.replace(Suppressed_val, np.nan, inplace=True)

# Also catch cells that are just whitespace
df = df.map(lambda x: np.nan if isinstance(x, str) and x.strip() == "" else x)

print(" Suppressed values replaced with NaN.")
print(f"\nTotal NaN count across dataset: {df.isna().sum().sum():,}")
```

```python
ID_cols = ["state", "district", "ruralurban_classification"]

# Check which ID cols actually exist
ID_cols = [c for c in ID_cols if c in df.columns]
print(f"Identifier columns kept as-is: {ID_cols}")

# All remaining columns  numeric
feature_cols = [c for c in df.columns if c not in ID_cols]

for col in feature_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

print(f"\n Converted {len(feature_cols)} columns to numeric.")
print(f"Remaining non-numeric (ID) columns: {ID_cols}")
```

```python
missing = (df[feature_cols].isna().sum().reset_index().rename(columns={"index": "column", 0: "missing_count"}))
missing["missing_pct"] = (missing["missing_count"] / len(df) * 100).round(2)
missing = missing.sort_values("missing_pct", ascending=False)

print(" Missing Value Summary (top 20):")
print(missing.head(20).to_string(index=False))
```

```python
# Droping colunms
col_missing_threshold  = 0.40   # Drop columns missing >40% of values
row_missing_threshold  = 0.40   # Drop rows (districts) missing >40% of values

cols_before = len(df.columns)
col_missing_frac = df[feature_cols].isna().mean()
cols_to_drop = col_missing_frac[col_missing_frac > col_missing_threshold].index.tolist()

df.drop(columns=cols_to_drop, inplace=True)
feature_cols = [c for c in feature_cols if c not in cols_to_drop]

print(f"Columns dropped (>{col_missing_threshold*100:.0f}% missing): "
      f"{len(cols_to_drop)}  {cols_to_drop}")

# Droping rows
rows_before = len(df)
row_missing_frac = df[feature_cols].isna().mean(axis=1)
df = df[row_missing_frac <= row_missing_threshold].copy()
df.reset_index(drop=True, inplace=True)

print(f"\nRows dropped  (>{row_missing_threshold*100:.0f}% missing): "
      f"{rows_before - len(df)}")
print(f"\n Dataset shape after dropping: {df.shape}")
```

```python
print(f"   Columns: {df.columns.tolist()}")
```

```python
df_clean = df

# Drop all urban/rural duplicate columns (suffixed _2, _3)
cols_to_keep = [c for c in df_clean.columns if not c.endswith(('_2', '_3'))]
df_clean = df_clean[cols_to_keep].copy()

print(f"Columns after  : {df_clean.shape[1]}")
print(f"Columns: {df_clean.columns.tolist()}")
print(f"Rows           : {df_clean.shape[0]}")
```
